```
l = [[1, 2, 3], [0, 4, 6], [7, 5, 8]]
gl = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
import copy

def index2d(list2d, value):
    return next((i, j) for i, lst in enumerate(list2d) for j, x in enumerate(lst) if x == value)


def up(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m - 1][n]
    li[m - 1][n] = temp
    return li


def down(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m + 1][n]
    li[m + 1][n] = temp
    return li


def right(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m][n + 1]
    li[m][n + 1] = temp
    return li


def left(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m][n - 1]
    li[m][n - 1] = temp
    return li

def puzzle(il, glt):
    if il == glt:
        print(il)
    else:
        i, j = index2d(il, 0)
        if i == 0 and j == 0:
            puzzle((right(il, i, j)), glt)
            puzzle((down(il, i, j)), glt)
        elif i == 0 and j == 1:
            puzzle((left(il, i, j)), glt)
            puzzle((right(il, i, j)), glt)
            puzzle((down(il, i, j)), glt)
        elif i == 0 and j == 2:
            puzzle((left(il, i, j)), glt)
            puzzle((down(il, i, j)), glt)
        elif i == 1 and j == 0:
            puzzle(up(il, i, j), glt)
            puzzle(right(il, i, j), glt)
            puzzle(down(il, i, j), glt)
        elif i == 1 and j == 1:
            puzzle(up(il, i, j), glt)
            puzzle(left(il, i, j), glt)
            puzzle(right(il, i, j), glt)
            puzzle(down(il, i, j), glt)
        elif i == 1 and j == 2:
            puzzle(up(l, i, j), glt)
            puzzle(left(l, i, j), glt)
            puzzle(down(l, i, j), glt)
        elif i == 2 and j == 0:
            puzzle(up(l, i, j), glt)
            puzzle(left(l, i, j), glt)
            puzzle(down(l, i, j), glt)
        elif i == 2 and j == 1:
            puzzle(up(l, i, j), glt)
            puzzle(left(l, i, j), glt)
            puzzle(down(l, i, j), glt)
        elif i == 2 and j == 2:
            puzzle(up(l, i, j), glt)
            puzzle(left(l, i, j), glt)
            puzzle(down(l, i, j), glt)


puzzle(l, gl)

```
