```
l = [[1, 2, 3], [0, 4, 6], [7, 5, 8]]
gl = [[1, 2, 3], [4, 5, 6], [7, 8, 0]]
import copy

def index2d(list2d, value):
    for i in range(3):
        for j in range(3):
            if list2d[i][j] == value:
                return i,j


def up(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m - 1][n]
    li[m - 1][n] = temp
    return li


def down(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m + 1][n]
    li[m + 1][n] = temp
    return li


def right(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m][n + 1]
    li[m][n + 1] = temp
    return li


def left(L, m, n):
    li = copy.deepcopy(L)
    temp = li[m][n]
    li[m][n] = li[m][n - 1]
    li[m][n - 1] = temp
    return li

def puzzle(l, gl):
    il = [l]
    k = 0
    n = 0
    t = 0
    while gl not in il:
        k = len(il) - t - 1
        length = len(il)
        print(k,length)
        for v in range(k, length):
            t = 0
            i, j = index2d(il[v], 0)
            if i == 0 and j == 0:
                il.append(right(il[v], i, j))
                il.append(down(il[v], i, j))
                t += 3
            elif i == 0 and j == 1:
                il.append(left(il[v], i, j))
                il.append(right(il[v], i, j))
                il.append(down(il[v], i, j))
                t += 4
            elif i == 0 and j == 2:
                il.append(left(il[v], i, j))
                il.append(down(il[v], i, j))
                t += 3
            elif i == 1 and j == 0:
                il.append(up(il[v], i, j))
                il.append(right(il[v], i, j))
                il.append(down(il[v], i, j))
                t += 4
            elif i == 1 and j == 1:
                il.append(up(il[v], i, j))
                il.append(left(il[v], i, j))
                il.append(right(il[v], i, j))
                il.append(down(il[v], i, j))
                t += 5
            elif i == 1 and j == 2:
                il.append(up(il[v], i, j))
                il.append(left(il[v], i, j))
                il.append(down(il[v], i, j))
                t += 4
            elif i == 2 and j == 0:
                il.append(up(il[v], i, j))
                il.append(left(il[v], i, j))
                t += 3
            elif i == 2 and j == 1:
                il.append(up(il[v], i, j))
                il.append(left(il[v], i, j))
                il.append(right(il[v], i, j))
                t += 4
            elif i == 2 and j == 2:
                il.append(up(il[v], i, j))
                il.append(left(il[v], i, j))
                t += 3
            n += 1
            print(n,il)

    index = il.index(gl)
    print("GOAL STATE REACHED IN ", index, " STEPS")
    print(il[index])


puzzle(l, gl)
```
